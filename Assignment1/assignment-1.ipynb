## Apply left join on `survey` data frame to merge it with `countries` using `dplyr` package

Your goal is to get the latitude and longitude columns onto the survey data, so you need to left join `survey` with `countries`. 

Caution: You may need to alter some country names either on the `countries` or `survey` data frames. 

You may want to go over the documentation of [dplyr](https://cran.r-project.org/web/packages/dplyr/vignettes/dplyr.html) package.

Tip: You may look into the [`left_join`](https://dplyr.tidyverse.org/reference/join.html) function of `dplyr`.

```r
countries = dplyr::rename(countries, country = country_name)
head(countries)
```

```r
##We should find unmatching country by comparing survey and country datasets. Operations for this could be seen below. 
distinct_country_in_survey = survey %>% distinct(country) %>% select(country)
distinct_country_in_country = countries %>% distinct(country) %>% select(country)
anti_join(distinct_country_in_survey, distinct_country_in_country, by="country")
```

```r
#Another way of doing the operation above(Finding ummatching country names)
anti_join(survey, countries, by="country")
```

```r
#Note: As could be seen in the previous step, unmatching country name is "United States of America"
#Note2: As the next step, we replace "United States of America" into "United States" as in survey database the matching word is "United States".
#These operations could be seen in the next step below.
```

```r
survey = survey %>% mutate(country=recode(country, "United States of America" = "United States"))
survey %>% distinct(country) %>% select(country)
```

```r
#Before joining survey and country data, let's do if there is any NA records in our suvey dataset. 
filter(survey, is.na(mental_health))
```

```r
#As you can see in the previous step, there are 55 NA records in survey dataset in mental_health colum. 
#As this column is an important attribute for the next steps, we eliminate this records as a preprocessing step.
```

```r
#Eliminating missing values in survey data
survey = filter(survey, !is.na(mental_health))
#we check again by filtering the same query
filter(survey, is.na(mental_health))
```

```r
#in the previous step, we eliminated mentalh_health records having NA valus. 
#Now we do left join as our next step.
```

```r
#...
survey = left_join(survey, countries, by = 'country', copy = FALSE)
head(survey)
```

```r
######### A quick explanation ###########

As a first step; I seperated the dataset into 3 groups based on gender. (survey_male, survey_female, survey_other)
As a second step; I find the averages(mean) of mental health depending of each group (male, female, other), based on the organization_size.
As a third step; I added a new column called gender into all these 3 datasets
As a fourth step; I unioned all these 3 data, by using funion function in data.table package
As a fifth step; I putted all the calculated numbers correctly into the ggplot function as input parameters
```

```r
#FIRST STEP
library(data.table)

survey_male = filter(survey, gender == '0')
survey_female = filter(survey, gender == '1')
survey_other = filter(survey, gender == '2')
```

```r
#SECOND STEP FOR MALE
#Average of mental_health per organization_size for male group
dt = data.table(survey_male)
avg_mental_health_by_male = dt[,mean(mental_health), organization_size];
avg_mental_health_by_male = dplyr::rename(avg_mental_health_by_male, AvgValue = V1)
avg_mental_health_by_male
```

```r
#SECOND STEP FOR FEMALE
#Average of mental_health per organization_size for female group
dt = data.table(survey_female)
avg_mental_health_by_female = dt[,mean(mental_health), organization_size];
avg_mental_health_by_female = dplyr::rename(avg_mental_health_by_female, AvgValue = V1)
avg_mental_health_by_female
```

```r
#THIRD STEP FOR OTHER GENDER GROUP
#Average of mental_health per organization_size for other group
dt = data.table(survey_other)
avg_mental_health_by_other = dt[,mean(mental_health), organization_size];
avg_mental_health_by_other = dplyr::rename(avg_mental_health_by_other, AvgValue = V1)
avg_mental_health_by_other
```

```r
#FOURTH STEP
avg_mental_health_by_male = avg_mental_health_by_male %>% mutate(gender= 'male')
avg_mental_health_by_female = avg_mental_health_by_female %>% mutate(gender= 'female')
avg_mental_health_by_other = avg_mental_health_by_other %>% mutate(gender= 'other')

avg_mental_health_by_male = data.table(avg_mental_health_by_male)
avg_mental_health_by_female = data.table(avg_mental_health_by_female)
avg_mental_health_by_other = data.table(avg_mental_health_by_other)


data = funion(avg_mental_health_by_male, avg_mental_health_by_female, all = FALSE)
data = funion(data, avg_mental_health_by_other, all = FALSE)
data
```
